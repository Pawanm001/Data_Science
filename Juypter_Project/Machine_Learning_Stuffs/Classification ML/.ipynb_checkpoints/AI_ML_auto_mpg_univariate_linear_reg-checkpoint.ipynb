```python
# Load csv using pandas
import numpy
import urllib
import pandas

#read csv file using pandas
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']

#independent variable chosen is horsepower. If any other independent 
#variable suits better than horsepower, it can be repleaced with that.
#target variable is mpg
usecols = ['mpg', 'horsepower']
datatype = {'mpg': numpy.float64}

dataset = pandas.read_csv(url, delim_whitespace=True, names=names, usecols=usecols, dtype=datatype)

print dataset
dataset = dataset[dataset.horsepower != '?']

dataset[['horsepower']] = dataset[['horsepower']].apply(pandas.to_numeric)

X = dataset.drop('mpg', axis=1)
y = dataset[['mpg']]


from sklearn import model_selection
validation_size = 0.20
seed = 7
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X, y, 
                      test_size=validation_size, random_state=seed)


from sklearn.linear_model import LinearRegression

regression_model = LinearRegression()
regression_model.fit(X_train, Y_train)

print "regression_model.coef: ", regression_model.coef_[0]
print "regression_model.intercept: ", regression_model.intercept_

b = regression_model.intercept_
m = regression_model.coef_[0]


y_predict = regression_model.predict(X_validation)

accuracy  = regression_model.score(X_validation, Y_validation)

print "Accuracy of model: ", accuracy

from sklearn.metrics import mean_squared_error

regression_model_mse = mean_squared_error(y_predict, Y_validation)

print "regression_model mean squared error: ", regression_model_mse

import math

print "so mean error: ", math.sqrt(regression_model_mse)

#plot points and best fit line.

import matplotlib.pyplot as plt
%matplotlib inline

X_train = X_train.reset_index().values
X_validation = X_validation.reset_index().values
Y_validation = Y_validation.reset_index().values
Y_train = Y_train.reset_index().values

for i in range(0, len(X_train)):
        x = X_train[i, 1]
        y = Y_train[i, 1]
        plt.scatter(x,y,color='black')
        
for i in range(0, len(X_validation)):
        x = X_validation[i, 1]
        y = Y_validation[i, 1]
        plt.scatter(x,y,color='red')

#best fit line
plt.plot(dataset.horsepower, (m * dataset.horsepower  + b), color='blue',linewidth=3)

```
